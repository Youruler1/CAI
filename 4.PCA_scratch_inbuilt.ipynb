```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_scaled=ss.fit_transform(X)
print(X_scaled)
```

#mean_: The mean of each feature, used to center the data.

#scale_: The standard deviation of each feature, used for scaling.

```python
ss.mean_
```

```python
ss.scale_
```

```python
data_original = ss.inverse_transform(X_scaled)
print(data_original)
```

```python
#3. Find covariance matrix
cov=np.cov(X_scaled,rowvar=False)
```

```python
print(cov.shape)
```

```python
#4. find eigenvalues and eigenvectors
values,vectors=np.linalg.eig(cov)
print(values,vectors)
```

```python
#5. Find expected variance
explained_variance=[]
for i in range(len(values)):
    explained_variance.append((values[i]*100)/np.sum(values))
print(explained_variance)
```

```python
#6. Compute Transformed dataset
column1=X_scaled.dot(vectors.T[0])
print(column1)
column2=X_scaled.dot(vectors.T[1])
res=pd.DataFrame(data=column1,columns=['PC1'])
res['PC2']=column2
res['species']=Y
print(res)
```

```python
sns.heatmap(res.iloc[:,:2].corr(),annot=True)
```

```python
res.plot(kind='scatter',x='PC1',y='PC2')
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
print("Explained Variance:", pca.explained_variance_ratio_)
print("Eigen vectors:\n", pca.components_) #eigenvectors sorted by the amount of variance they capture.
```
