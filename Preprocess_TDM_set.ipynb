```python
content=['data science important field science',
 'important data science course',
 'cars driven roads',
 'trucks driven highways']
```

```python
def extract_vocabulary(corpus):
  vocab=set()
  for sentence in corpus:
    for word in sentence.split():
      vocab.add(word)
  return (sorted(vocab))
  #print(vocab)

```

```python
vocab=extract_vocabulary(content)
```

#DTM

```python
def create_DTM(corpus,vocab):
  dtm=[]
  for sentence in corpus:
    row=[0]*len(vocab)
    for word in sentence.split():
      i=vocab.index(word)
      if i != -1:
        row[i]+=1
    dtm.append(row)
  return dtm

```

```python
print(create_DTM(content,vocab))
```

```python
import pandas as pd
df=pd.DataFrame(create_DTM(content,vocab),columns=vocab)
df
```

#Binary DTM

```python
def create_BinaryDTM(corpus,vocab):
  dtm=[]
  for sentence in corpus:
    row=[0]*len(vocab)
    for word in sentence.split():
      i=vocab.index(word)
      if i != -1:
        row[i]=1
    dtm.append(row)
  return dtm
```

```python
import pandas as pd
df=pd.DataFrame(create_BinaryDTM(content,vocab),columns=vocab)
df
```

```python
# Initialize an empty list to hold the matrix
import pandas as pd

def create_TDM(content,vocab):
  tdm = []
  # Iterate over each term in unique_terms
  for term in vocab:
      # Create a row for the current term
      row = []
      # Check the presence of the term in each document
      for doc in content:
          if term in doc.split():
              row.append(1)
          else:
              row.append(0)
      # Append the row to the TDM
      tdm.append(row)
   # Convert the TDM to a DataFrame for better visualization (optional)

  tdm_df = pd.DataFrame(tdm, index=vocab)
  print(tdm_df)
```

#CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample dataset
documents = [
    "Machine learning is fascinating.",
    "Data science and machine learning go hand in hand.",
    "Deep learning is a subset of machine learning.",
    "Python is great for machine learning.",
    "Data science is about extracting knowledge from data."
]

# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the documents
X = vectorizer.fit_transform(documents)

# Get feature names (i.e., the vocabulary)
feature_names = vectorizer.get_feature_names_out()
#vocabulary_: Returns a dictionary mapping features to indices.
vocab = vectorizer.vocabulary_

# Convert the result to an array
X_array = X.toarray()

# Print the results
print("Feature Names:", feature_names)
print("Vocab:", vocab)

print("Document-Term Matrix:")
print(X_array)
```

#TFIDF from scratch

```python
content=['data scienc import field scienc',
 'thi import data scienc cours',
 'the car driven road',
 'the truck driven highway']
```

```python
vocab=extract_vocabulary(content)
```

```python
d={}
for term in vocab:
  d[term] = 0
  for sentence in content:
    if term in sentence.split():
      d[term] += 1


```

```python
import numpy as np
idf={}
for term in vocab:
  idf[term]=np.log10(len(content)/d[term])
  print(idf[term])
```
